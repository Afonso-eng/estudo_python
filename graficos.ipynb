grafico de barras empilhadas

```python
import matplotlib.pyplot as plt
import numpy as np
anos=[2010,2011,2012,2013,2014,2015,2016,2017,2018,2019,2020]
orcados=[100,100,100,100,100,100,100,100,100,100,100]
real=[150,10,100,80,120,50,23,91,100,400,-70]
plt.bar(anos,orcados,label="orcados")
plt.bar(anos,real,label="real") 
plt.title("orcados e real")
plt.xlabel("anos")
plt.ylabel("valores")
plt.legend()
plt.show()
```

grafico de linhas

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Definição dos meses e vendas
meses = ['Janeiro', 'Fevereiro', 'Março', 'Abril', 'Maio', 'Junho', 
         'Julho', 'Agosto', 'Setembro', 'Outubro', 'Novembro', 'Dezembro']
vendas = [85, 90, 78, 92, 88, 96, 94, 98, 100, 102, 105, 108]

# Criar figura
plt.figure(figsize=(17, 8))

# Criar o gráfico de linha com marcadores
plt.plot(meses, vendas, marker='o', linestyle='-', color='b', label='Vendas')

# Encontrar os índices do menor e maior valor
min_idx = np.argmin(vendas)
max_idx = np.argmax(vendas)

# Adicionar destaque ao menor valor
plt.scatter(meses[min_idx], vendas[min_idx], color='red', marker='o', s=100, label=f'Mínimo: {vendas[min_idx]}')

# Adicionar destaque ao maior valor
plt.scatter(meses[max_idx], vendas[max_idx], color='green', marker='o', s=100, label=f'Máximo: {vendas[max_idx]}')

# Adicionar rótulos em todos os pontos
for i in range(len(meses)):
    plt.text(meses[i], vendas[i] + 1, str(vendas[i]), fontsize=10, ha='center', color='black')

# Configurações adicionais
#plt.xlabel("Meses")
plt.ylabel('')
plt.title("Vendas Mensais")
plt.xticks(rotation=45)  # Inclina os nomes dos meses para melhor visualização
plt.yticks([])
plt.grid(True, linestyle="--", alpha=0.6)  # Adiciona um grid leve ao gráfico
plt.legend()  # Exibe a legenda

plt.axvspan(6,11, color='gray', alpha=0.1)
plt.text(6, 80, 'Recuperação de crescimento', fontsize=15, color='black', ha='right')
plt.savefig('grafico_VENDAS.png',bbox_inches='tight',dpi=300)
# Exibir gráfico
plt.show()
```

# grafico de barra personalizado

```python
calorias_por_alimneto ={
"Peito frango": 165,
"Filé peixe": 96 ,
"Carne bovina": 250 ,
"Carne suína": 280 ,
"Salmão grelhado": 206 
}
alimentos=list(calorias_por_alimneto.keys())
calorias=list(calorias_por_alimneto.values())
alimentos=[alimentos.replace(" ","\n") for alimentos in calorias_por_alimneto.keys()]
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 5))
barras=plt.bar(alimentos, calorias_por_alimneto.values(),color='grey',alpha=0.5)
barras[calorias.index(max(calorias))].set_color('darkred')  
barras[calorias.index(max(calorias))].set_alpha(0.8) 
plt.bar_label(barras,fmt='%.1f',fontweight='bold',color='white',padding=-15)
plt.box(False)
plt.yticks([])
plt.tick_params(axis='x',length=0)
plt.title('Calorias por 100g de alimento',color='grey',fontsize=20,fontweight='bold')
plt.savefig('grafico_calorias_alimentos.png',bbox_inches='tight',dpi=300)
plt.show()
```
