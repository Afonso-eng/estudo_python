```python
# criar um grafico de climograma
import matplotlib.pyplot as plt
meses=['JAN','FEV','MAR','ABR','MAI','JUN','JUL','AGO','SET','OUT','NOV','DEZ']
temp_media=[25.5,21.5,20.5,15.5,16.5,13.5,14.5,15.5,19.5,25.5,21.5,22.5]
precipitacao=[150,120,100,80,60,40,20,20,40,60,80,100]

fig, ax1 = plt.subplots(figsize=(8,5))


cor_temperatura='tab:red'
cor_precipitacao='tab:blue'

ax1.bar(meses,precipitacao,color=cor_precipitacao,label='precipita√ß√£o')
ax1.set_ylabel('precipita√ß√£o em MM',color=cor_precipitacao)
ax1.tick_params(axis='y',labelcolor=cor_precipitacao)





ax2 = ax1.twinx()
ax2.plot(meses,temp_media,color=cor_temperatura,linewidth=2,marker='o',label='temperatura')
ax2.set_ylabel('temperatura em ¬∞C',color=cor_temperatura)
ax2.tick_params(axis='y',labelcolor=cor_temperatura)



ax1.set_title('Climograma')
#adicionando legenda

fig.legend(loc="upper left", bbox_to_anchor=(0.4,1), bbox_transform=ax1.transAxes)

plt.show()
```

```python
import matplotlib.pyplot as plt
mosaico='AA;BB;CD'
fig = plt.figure(figsize=(12,10))
axs = fig.subplot_mosaic(mosaico)


#grafico de linha
meses=['jan','fev','mar','abr','mai','jun','jul','ago','set','out','nov','dez']
vendas=[1500,1727,1350,1775,1052,1200,1300,1200,1100,1400,1500,1800]
axs['A'].plot(meses,vendas)
#grafico de barras
produtos =['iphone','samsung','xiaomi','motorola']   
qtd = [15000,12000,14000,17000]
axs['B'].bar(produtos,qtd)
#grafico de pizza
categorias = ['Python', 'JavaScript', 'Java', 'C#', 'PHP']
quatidades = [25, 20, 15, 20, 10]
axs['C'].pie(quatidades, labels=categorias)
#grafico de barra empilhada
anos=[2015,2016,2017,2018,2019,2020,2021]
vendas_online=[100,200,300,400,500,600,700]
vendas_loja=[10,20,30,40,50,60,70]

axs['D'].bar(anos,vendas_loja, label='Vendas Loja',bottom=vendas_online)
axs['D'].bar(anos,vendas_online, label='Vendas Online')
axs['D'].set_title('Vendas Online e Loja')
axs['D'].set_xlabel('Anos')
axs['D'].set_ylabel('Vendas')

fig.suptitle('Gr√°ficos em Python',fontsize=16)

plt.show()
```

```python
meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
novos_usuarios = [100, 80, 110, 140, 150, 160, 170, 120, 110, 150, 200, 140]
usuarios_ativos_diarios = [1000, 750, 900, 1300, 1400, 1500, 1200, 1200, 1500, 1900, 2000, 1800]
receita_mensal = [10000, 8500, 9800, 10000, 11000, 12000, 10000, 10500, 13000, 13000, 20000, 17000]
tickets_suporte_abertos = [110, 75, 100, 140, 170, 160, 100, 130, 190, 200, 210, 140]
tickets_suporte_fechados = [100, 85, 100, 120, 130, 150, 140, 110, 130, 120, 150, 200]
```

```python
import matplotlib.pyplot as plt

mosaico='AA;BB;CD'
fig=plt.figure(figsize=(10,10))
espaco={'hspace':0.3,'wspace':0.3}
axs=fig.subplot_mosaic(mosaico,gridspec_kw=espaco)
#grafico 01
axs['A'].plot(meses,novos_usuarios)
axs['A'].set_title('Novos Usuarios por mes')
axs['A'].set_ylabel('novos usuarios')
#grafico 02
axs['B'].bar(meses,receita_mensal)
axs['B'].set_title('Receita por mes R$')
axs['B'].set_ylabel('receita')
#grafico 03
axs['C'].plot(meses,usuarios_ativos_diarios)
axs['C'].set_title('Usuarios Ativos por dia')
axs['C'].set_ylabel('m√©dia de usu√°rios ativos')
#grafico 04
axs['D'].bar(meses,tickets_suporte_abertos)
axs['D'].bar(meses,tickets_suporte_fechados,bottom=tickets_suporte_abertos)
axs['D'].set_title('Tickets de Suporte')
axs['D'].set_ylabel('tickets')
#ajuste dos graficos
fig.suptitle('Dashboard de Analise de Dados',fontsize=16,color='darkblue')
plt.show()
```
