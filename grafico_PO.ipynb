```python

import matplotlib.pyplot as plt
meses=['jan','fev','mar','abr','mai','jun','jul','ago','set','out','nov','dez']
vendas=[1500,1727,1350,1775,1052,1200,1300,1200,1100,1400,1500,1800]

fig, ax=plt.subplots()
ax.plot(meses,vendas,color='darkblue')
ax.set_xlabel('Meses')
ax.set_ylabel('Vendas')
ax.set_title('Vendas por m√™s')
plt.box(False)
plt.show()
```

# grafico de barras

```python
import matplotlib.pyplot as plt
prdutos =['iphone','samsung','xiaomi','motorola']   
vendas = [15000,12000,14000,17000]
fig, ax = plt.subplots()
ax.bar(prdutos,vendas)
ax.set_title('Vendas de celulares')
ax.set_xlabel('Produtos')
ax.set_ylabel('Vendas')
plt.show()
```

# GR√ÅFICO DE PIZZA

```python
import matplotlib.pyplot as plt
categorias = ['Python', 'JavaScript', 'Java', 'C#', 'PHP']
quatidades = [25, 20, 15, 20, 10]
destacados = [True, False, False, True, False]
fig, ax = plt.subplots()
ax.pie(quatidades, labels=categorias, autopct='%1.1f%%', explode=[0.1, 0, 0, 0, 0])
plt.show()

```

```python
anos=[2015,2016,2017,2018,2019,2020,2021]
vendas_online=[100,200,300,400,500,600,700]
vendas_loja=[10,20,30,40,50,60,70]

fig, ax = plt.subplots()
ax.bar(anos,vendas_loja, label='Vendas Loja',bottom=vendas_online)
ax.bar(anos,vendas_online, label='Vendas Online')
ax.set_title('Vendas Online e Loja')
ax.set_xlabel('Anos')
ax.set_ylabel('Vendas')
plt.legend()
plt.show()



```
